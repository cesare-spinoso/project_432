## Helper functions
The following functions will be heavily used throughout this notebook. We find it useful to create a separate section for them.

```python
# This function will be used to create the time series sequences used for training
def create_shift_sequence(data, seq_len, num_inputs=1, step_size=1):
  '''
  Returns multi-dimensional feature matrix X and label matrix Y specifically
  designed for time series data. This function was originally designed to be
  used only for recurrent networks, but by a simple transpose we can
  extend its functionality to also produce feature and label matrices X and Y
  for a feed-forward structure.
            
            Parameters:
                    data () : Sequence of raw input data. Expects a time 
                    series based data. Can be either a pandas series or a 1-D 
                    array (for example a 1-D tensor) in the case of uni-
                    variate time series. In the case of multivariate data
                    can be a pandas dataframe of any 2-D array (for example
                    a 2-D tensor).
                    seq_len (int) : The desired sequence length of the feature
                    matrix X.
                    num_inputs (int) : The number of corresponding features
                    the feature matrix X should. Defaults to 1 because the
                    default setting expects a univariate time series.
                    step_size (int) : In what increments is the time series data
                    taken. Defaults to 1.

            Returns:
                    X (tensor) : The feature matrix which has shape (seq_len,
                    num_batches, num_inputs).
                    Y (tensor) : The label matrix which has shape (seq_len, 
                    num_batches).
  '''
  # Because we want to cover all the data
  if (len(data) - 2*seq_len) % step_size != 0:
    raise ValueError("Your step_size is not compatible.")
  # X scans the data in runs of length seq_len and in increments of step_size
  # Y is the corresponding run of length seq_len which comes AFTER
  # the run  for X. As a result, we can't scan more than len(data) - 2*seq_len
  # Imagine scanning the data with [X1 ... Xseq_len] - - [Y1 ... Yseq_len]
  # with steps of step_size.
  num_batches = int((len(data) - 2*seq_len)/step_size + 1)
  X, Y = \
    torch.zeros(seq_len, num_batches, num_inputs), torch.zeros(seq_len, num_batches)
  # i is the index for the X and Y, j is the scanning index for the data
  for i, j in zip(range(0, num_batches), range(0, len(data) - 2*seq_len + 1, step_size)): 
      X[:, i, 0] = data[j:j+seq_len]
      Y[:, i] = data[j+seq_len:j+2*seq_len]
  return X.type(torch.float32), Y.type(torch.float32)
```

```python
# The following will be used for creating the validation set used to validate
# out-of-sample performance. The date is taken to be one full continuous
# sequence where the first seq_len point correspond to the sequence fed into
# the neural network and data[seq_len to 2*seq_len] is used as the corresponding label.
def create_val_set(data, seq_len):
  return data[:seq_len], data[seq_len:2*seq_len]
```

```python
# This function will be used to train our vanilla feed-forward neural network
def train_vanilla_net(X_trn, Y_trn, X_val, Y_val, val_size, batch_size, scaler,
                      num_epochs, learning_rate, optimizer, model, **kwargs):
    torch.manual_seed(0)
    # The assumption is that X_trn is given in (num_batches, num_features)
    # and Y_trn is given in (num_batches, output_size)
    # X_val should be a single batch and so should Y_val

    # Define model, loss function and optimizser (will be using Adam)
    model = model(**kwargs)
    loss = torch.nn.MSELoss()
    optimizer = optimizer(model.parameters(), lr=learning_rate)

    num_batches = X_trn.shape[0] # The first dimension is assumed to have number of batches

    # Standard pytorch training
    model.train()
    for epoch in range(1, num_epochs+1):
      for i in range(0, num_batches, batch_size):
          x = X_trn[i:i+batch_size, :] 
          y = Y_trn[i:i+batch_size, :]
          y_pred = model(x)
          l = loss(y_pred, y)
          model.zero_grad()
          l.backward()
          optimizer.step()
      if epoch % 10 == 0:
        print(f"Loss at epoch # {epoch}: {round(l.item(), 4)}")

    model.eval() # To turn dropout off
    with torch.no_grad():
      # Evaluate out-of-sample error, reshape into 2-D matrix for sklearn
      y_pred = model(X_val).reshape(-1, 1) 
      y_pred_unscaled = scaler.inverse_transform(y_pred)
      y_true = scaler.inverse_transform(Y_val.reshape(-1, 1))
      mse = mean_squared_error(y_pred_unscaled[:val_size], y_true[:val_size])
      rmse = math.sqrt(mse)
    return rmse, mse
```

```python
# This function will be used to train our recurrent neural networks
# We admit that the function is alost copy-pasted the function above,
# the difference is that the shape of X_trn and Y_trn are different. 
# No val_size or input scaler are also required.
# Because we wanted to keep the function somewhat understandable we 
# preferred not to modularize this further. 
def train_rnn_timeseries(X_trn, Y_trn, X_val, Y_val, val_size, batch_size, num_epochs,
                         learning_rate, optimizer, model, **kwargs):
    torch.manual_seed(0)
    # The assumption is that X_trn is given in seq_len, num_batches, num_inputs
    # and Y_trn is given in seq_len, num_bacthes
    # X_val should be a single sequence of seq_len and so should Y_val

    # Define model, loss function and optimizser (will be using Adam)
    model = model(**kwargs)
    loss = torch.nn.MSELoss()
    optimizer = optimizer(model.parameters(), lr=learning_rate)

    num_batches = X_trn.shape[1] # the second dimension is assumed to have number of batches

    # Standard pytorch training
    for epoch in range(1, num_epochs+1):
      for i in range(0, num_batches, batch_size):
          x = X_trn[:, i:i+batch_size, :] 
          y = Y_trn[:, i:i+batch_size]
          y_pred = model(x)
          l = loss(y_pred, y)
          model.zero_grad()
          l.backward()
          optimizer.step()
      if epoch % 10 == 0:
        print(f"Loss at epoch # {epoch} : {round(l.item(), 4)}")

    with torch.no_grad():
      # The X_val is assumed to be in 1-D, convert it to (seq_len, 1, 1) for 
      # the model to accept the input. 
      y_pred = model(X_val.view(-1, 1, 1)).reshape(-1, 1) 
      y_pred_unscaled = scaler.inverse_transform(y_pred)
      y_true = scaler.inverse_transform(Y_val.reshape(-1, 1))
      mse = mean_squared_error(y_pred_unscaled[:val_size], y_true[:val_size])
      rmse = math.sqrt(mse)
    return rmse, mse

  
```

### Time Series Neural Network
We explore using vanilla neural networks (feed-forward neural network with variable number of hidden layers and hidden units). To account for time series data we train a first model using a $365 \times 24$ - hidden layers - $365 \times 24$ network where the input uses a year's worth of hourly time series observations as features (1 feature per hour) and the output is trained on the following year's worth of hourly data. We used a year's worth of annual data rather than half a year (which relates to our primary goal) because it did not pose much computational overhead. The validation was done simply by selecting the first half of the output (from the prediction on the validation set).

```python
# Vanilla NNet module, allow variable number of layers with the same activation
class VanillaNNet(torch.nn.Module):
      def __init__(self, input_size, num_layer_list, output_size, activation, dropout=None):
        super(VanillaNNet, self).__init__()
        if len(num_layer_list) < 1:
          raise ValueError("No hidden layers defined.")
        # List to iterate over
        hidden_output_list = num_layer_list + [output_size]
        # Create module list, sequence of hidden layers with given activation
        self.hidden = torch.nn.ModuleList([torch.nn.Linear(input_size, num_layer_list[0])])
        # In case of dropout on the input units 
        if dropout != None:
          self.hidden.extend([torch.nn.Dropout(dropout[0])])

        for first_dim, second_dim in zip(hidden_output_list, hidden_output_list[1:]):
          if dropout == None:
            self.hidden.extend([activation,
                                torch.nn.Linear(first_dim, second_dim)])
          else:
            self.hidden.extend([activation,
                                torch.nn.Dropout(dropout[1]),
                                torch.nn.Linear(first_dim, second_dim)])
      
      def forward(self, x):
        for layer in self.hidden:
          x = layer(x)
        return x  
```

As discussed in the documentation for `create_shift_sequence`, the $X$ we feed into our neural network consists of a sequences of the time series. For this feed-forward network, input $X = [X_1, X_2 \dots, X_{365\times24}]$ where $X_t$ corresponds to the $t$-th time series observation and the corresponding label is $Y = [X_{365\times24 +1}, \dots, X_{2\times365\times24}]$ i.e. the time series values for the next year. This is how the network is trained. It is validated using the first-half of the 2017 time series data so as to faithfully estimate the test-time performance metric.



```python
# Define the training and validation set
# The sequence length is set to a year to compare with other models in the validation
# that is we use year's worth of hourly data as the window.
seq_len = 365*24 
# We use a stepsize of 24 to reduce the batch size so that every new X, Y corresponds
# to a shift by 24 hours.
X_trn, Y_trn = create_shift_sequence(ts_train, seq_len, step_size=24)
# Transpose so that the data can be used by the feed-forward vanilla network
X_trn = torch.transpose(X_trn.reshape(seq_len, -1), 0, 1) 
Y_trn = torch.transpose(Y_trn.reshape(seq_len, -1), 0, 1)
# The validation set uses the last year as the input and the next year as the label
X_val, Y_val = create_val_set(torch.cat((ts_train[-seq_len: ], ts_val), 0), seq_len=seq_len)
```

### Introduce features
We now see if we can add features to the model to explain the remaining error. That is if we have $y = \hat{y} + \epsilon$ where $y$ is the true value of teh time series and $\hat{y}$ is the predicted value using the 8760 - 700 - 8760  neural network, then we use another neural network using the extracted features to predict $\epsilon$. We choose this approach become we believe that it is a more natural and also simpler implementation than if we had added to the 8760 - 700 - 8760 the 11 features for each observation (i.e. a neural network with input size 8760 + $8760 \times 11$ ). This simplication comes at the loss of the potential interatction that the features would have with the pervious time series observations.

We now go through the process of finding the residuals for the time series neural network. We will use the network to predict the electricity demand for 2013 to 2016 inclusively. We do this by taking the year 2013 and predicting 2014, using 2014 to predict 2015 and so on. We do this as it is the simplest way of utilizing the previous model to find all the errors. We do not claim this to be the best approach, but it is convenient given the way we set up the model to predict a whole year's sequence. 

```python
# We cut out the first two days to create sequences of 8760 that fit within the
# trainig set.
cut_out = 2*24
seq_len = step_size = 365*24
X_trn, Y_trn = create_shift_sequence(ts_train[cut_out:], seq_len, step_size=step_size)
X_trn = torch.transpose(X_trn.reshape(input_size, -1), 0, 1) 
Y_trn = torch.transpose(Y_trn.reshape(input_size, -1), 0, 1)
```

```python
# Create the list of residuals
Y_res_trn = []
# Create a new scaler object to scale the residuals
scaler_dff = MinMaxScaler(feature_range=(-1, 1))
# Compute residuals in original scale and then scale them
with torch.no_grad():
  for x, y in zip(X_trn, Y_trn):
    y_pred = model(x)
    # Unscale because scaling is not closed under difference
    y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
    y = scaler.inverse_transform(y.reshape(-1, 1))
    diff = scaler_dff.fit_transform(y - y_pred)
    diff = torch.tensor(diff, dtype=torch.float32)
    Y_res_trn.append(diff)
# These are the targets that we will use for teh features
Y_res_trn = torch.cat(Y_res_trn)
```

```python
# Get the validation set that will be used the validate the feature training
X_val, Y_val = create_val_set(torch.cat((ts_train[-seq_len: ], ts_val), 0), seq_len=seq_len)
with torch.no_grad():
  # Again we unscale and then rescale using the new scaler object for the difference
  y = scaler.inverse_transform(Y_val.reshape(-1, 1))
  y_pred = scaler.inverse_transform(model(X_val).reshape(-1, 1))
  Y_res_val = (y - y_pred).reshape(-1, 1)
  Y_res_val = scaler_dff.transform(Y_res_val)
  Y_res_val = torch.tensor(Y_res_val[:VAL_SIZE], dtype=torch.float32)
```

The neural network will have input size equal to the number of features that have been extracted and will predict for the corresponding residual (i.e. at that given time). Thus this will be a 10 - $x$ - 1 neural network. We will tune the hidden layers $x$, the batch size and we will also introduce dropout as a regularizer. This is because we do not want to consider an exhaustive search on the feature space like we did in linear models both for computational reasons and because dropout is a much better suited regularizer. We train on both no dropout and a dropout of 0.2 for the input units (i.e. $p=0.2$ that one of the input units is dropped) and of 0.5 for the hidden units as per [Hinton et al.](https://arxiv.org/pdf/1207.0580.pdf).

## LSTM
In this section we try to use a more sequence based neural network to adapt to our time series data. We do this with the help of recurrent networks particularly LSTMs. We do not forsee a vaniall RNN to provide interesteing results since the length of the sequence that we will use to train the recurrent network will be long (hourly half-year data and potentially yearly). So we immediately go to LSTMs. For this section as well as GRUs we relied on the following: [source 1](https://stackabuse.com/time-series-prediction-using-lstm-with-pytorch-in-python/) [source 2](https://www.jessicayung.com/lstms-for-time-series-in-pytorch/).

```python
# Create LSTM model
class TimeSeriesLSTM(torch.nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size=1, 
                 dropout=0,
                 bidirectional=False):
        super(TimeSeriesLSTM, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.bidirectional = bidirectional
        self.lstm = torch.nn.LSTM(input_size=input_size,
                                  hidden_size=hidden_size,
                                  num_layers=num_layers,
                                  bidirectional=bidirectional,
                                  dropout=dropout)
    
    # Reference this code here
    def forward(self, X):
        lstm_out, _ = self.lstm(X)
        if self.bidirectional:
          return lstm_out[:, :, 0].view(X.shape[0], -1) # take the forward direction
        else:
          return lstm_out.view(X.shape[0], -1)
    
```

The ideal way to split our data would be similar to the previous vanilla neural-network where we have a sequence of $24 \times 365$ i.e. We show the LSTM a year's worth of hourly data and train it to predict the next year's sequence. We can then validate it on the first half of the hourly data for 2017 (giving it the data from 2016).

## GRU
We also try the same search used for the LSTM but this time on GRUs. The idea is that the GRUs simpler architecture may increase computation speeds.

```python
# Create GRU model
class TimeSeriesGRU(torch.nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size=1, # batch_size
                 dropout=0,
                 bidirectional=False):
        super(TimeSeriesGRU, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        # self.batch_size = batch_size 
        self.bidirectional = bidirectional
        self.gru = torch.nn.GRU(input_size=input_size,
                                  hidden_size=hidden_size,
                                  num_layers=num_layers,
                                  bidirectional=bidirectional,
                                  dropout=dropout)
    
    # Reference this code here
    def forward(self, X):
        gru_out, _ = self.gru(X)
        if self.bidirectional:
          return gru_out[:, :, 0].view(X.shape[0], -1) # take the forward direction
        else:
          return gru_out.view(X.shape[0], -1)
    
```

```python
# Set up the hyperparameter search
layer_list = list(range(2, 5))
batch_list = [183, 549]
bidirectional = [False, True]
dropout = [0.0] # double number of hidden units

model_selection = list(itertools.product(*[batch_list, layer_list, bidirectional, dropout]))
```

```python
# Set up the training parameters
num_epochs = 75
optimizer = torch.optim.Adam
model = TimeSeriesGRU
```
