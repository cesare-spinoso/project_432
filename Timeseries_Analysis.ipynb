## Training
A note before we get started on training which will hold for the rest of this notebook and *all other notebooks* we will be performing 10-fold cross-validation in most of our training phases. This translates to validation score that roughly represents how well our model could predict electricity demand half a year into the future (since our training set is 5 years and we have 10 folds). For training methods where we will not use 10-fold (neural networks) and rather just take a validation set, it will be for half a year as well to keep everything consistent. We will use the mse and rmse as our main training metric. When we report results at the end we may also provide more interpretable metrics.

Training in all the models in this notebook will be done on 2012-2016 and will be validated via cross-validation. We will then retrain the optimal model on 2012-2016 and then plot it on the first half of 2017 to understand its behaviour (the same plot interval will be done for all of our models).

### Baseline estimators
We use mean, median and lag estimators as our baselines computing the mse and rmse validation scores.

```python
# Define a function to apply k_fold cross validation with n_splits on stricitly 
# time_series data i.e. don't expect any other features but time. 
# This was necessary because typically sklearn expects X and y for a regression
# but for time series X and y are from the same column.
def rolling_window_time_series(model, n_splits, X, y):
    '''
    Returns the mse and rmse of the 'rolling window' cross validation

            Parameters:
                    model : A time series model
                    n_splits (int): Number of splits to perform
                    X (Series): The time series on which to train
                    y (Series): The time series on which to predict. May be the
                    same as X.

            Returns:
                     mse_cv (float), rmse_cv (float): Validation mse and rmse
    '''
    # Keep a running total of the mean mse
    mse_cv = 0
    N = 1
    ts_splitter = sklearn.model_selection.TimeSeriesSplit(n_splits=n_splits)
    for train, valid in ts_splitter.split(X):
        # For each train, valid fold fit it on train
        model.fit(X[train], y[train])
        # Predict on valid
        y_pred = model.predict(X[valid])
        # Find mse of this fold and then add it to the running average
        mse_fold = sklearn.metrics.mean_squared_error(y_true=y[valid], y_pred=y_pred)
        # This is to keep a running average (instead of appending to an array 
        # and then taking the mean)
        if N == 1:
            mse_cv = mse_fold
        else:
            # This can be checked by expanding the formula for the mean
            mse_cv += (mse_fold - mse_cv)/N
        N+=1
    # Return mse and rmse
    return mse_cv, math.sqrt(mse_cv)
        
```

```python
# Dummy estimators that will be used as baseline for all more 'sophisticated' models.
dummy_mean = sklearn.dummy.DummyRegressor(strategy="mean")
dummy_median = sklearn.dummy.DummyRegressor(strategy="median")
```

```python
# We will consider a k-fold cross validation procedure where we make a fold about half a year's worth of data
# Since the mean and median's are fast to compute this does not pose any computational difficulty
n_splits = 10
# Apply rolling window cross-validation to the dummy mean mse
dummy_mean_mse, dummy_mean_rmse = rolling_window_time_series(dummy_mean,
                                                             n_splits,
                                                             ts_trn.index,
                                                             ts_trn.values)
dummy_median_mse, dummy_median_rmse = rolling_window_time_series(dummy_median,
                                                               n_splits,
                                                               ts_trn.index,
                                                               ts_trn.values)
```

```python
# What are the baselines?
print(f"The dummy validation mean mse is {round(dummy_mean_mse, 2)} and "
      f"the validation rmse is {round(dummy_mean_rmse, 2)}")
print(f"The dummy median validation mse is {round(dummy_median_mse, 2)} and "
      f"the validation rmse is {round(dummy_median_rmse, 2)}")
```

```python
# Create another custom dummy regressor class with two things: 
# 1. predicts X_t as its kth lag i.e. X_{t-k}. What this is actually doing is LOOCV 
# (but for time series).
# 2. if given X[1...t] then it predicts X[t+1...t+m] using X[t+1-m...t]
class CustomDummyForecastor:
    """
    A class to represent a custom dummy 'lagged' forecastor.

    ...

    Attributes
    ----------
    lag : int
      Indicates what time lag the forecastor will use for prediction.
    ts : Series
      Holds the time series to fit.
    ts_fit : Series
      Holds the fitted time series.

    Methods
    ----------
    fit(X, y):
      Fits the data
    predict(X):
      Make a lag 1 estimate.
    predict_all():
      Make a prediction for the lag passed to the instance.
    score():
      Score the prediction with mse and rmse.
    """
    
    def __init__(self, lag):
      """
      Constructs the lag attributes.

      Parameters
      ----------
      lag : int
        Indicates what time lag the forecastor will use for prediction.
      """
      self.lag = lag    
    
    # X is here as a placeholder (to use for cv function), pass the whole 
    # time series to y
    def fit(self, X, y):
      """
      Fit the passed time series simply by storing the passed time series in y.

      Parameters
      ----------
      X : 
        Does not expect anything to actually be passed here.
        Exists for the rolling_window function.
      y : Series
        Time series instance passed to the model.

      Returns
      -------
      None
      """
      self.ts = y
        
    def predict(self, X):
      """
      Predict the lag estimate of the time series.

      Parameters
      ----------
      X : 
        Inputted time series

      Returns
      -------
      Predicted lag estimate based on the length of X. That is, we will
      make the cross-validation procedure implicitly pick the lag.
      """
      self.ts_fit = self.ts[X.index - datetime.timedelta(hours=len(X))]
      return self.ts_fit
    
    def predict_all(self):
      """
      Predict the estimate of the time series with the lag.

      Parameters
      ----------
      None

      Returns
      -------
      Predicted estimate for given lag.
      """
      self.ts_fit = self.ts.shift(self.lag)
      return self.ts_fit
    
    # Returns the mse and rmse
    def score(self):
      """
      Return the mse and rmse of the given dummy forecastor.

      Parameters
      ----------
      None

      Returns
      -------
      Mse and rmse.
      """
      # Drop nas to use sklearn metrics
      y_true = self.ts[self.ts_fit.notna()]
      y_pred = self.ts_fit.dropna(axis=0)
      mse = sklearn.metrics.mean_squared_error(y_true=y_true, y_pred=y_pred)
      return mse, math.sqrt(mse)
    
```

```python
# Train and validate for multiple lags
# Use lags that we think might be useful: previous hour, previous day, week, month and year
# These are rough estimates since months and years (leap years) have a different number of hours.
lag_dic = {'hourly':1, 'daily':24, 'weeky':7*24, 'monthly':24*30, 'yearly':365*24}
mse_dic = dict()
for k, v in lag_dic.items():
    dummy_single_lag = CustomDummyForecastor(v)
    dummy_single_lag.fit(_, ts_trn) # don't use X for now
    dummy_single_lag.predict_all()
    mse_dic[k] = dummy_single_lag.score()
# Print the results
for k, v in mse_dic.items():
    print(f"The {k} lag estimate has validation mse {round(v[0], 2)} and "
         f"validation rmse {round(v[1], 2)}")
```

We see that using a naive estimator that just predicts the values from before gives an RMSE of 263 which is pretty good! It's a good thing we set up this baseline, otherwise we might have been too happy with an rmse of say 1000...But note that this is <b>VERY</b> misleading because remember our objective is to predict electricity demand for the next half a year.

```python
# Plot the best dummy estimator versus the true time series curve for a given 
# month (since the yearly plot will be too dense to see anything anyway)
# Chose a time interval
time_start = '2015-06-01'
time_end = '2015-06-08'
# Retrain the lag_1 estimator
dummy_lag_1 = CustomDummyForecastor(1)
dummy_lag_1.fit(_, ts_trn)
y_pred_lag_1 = dummy_lag_1.predict_all() 
# Plot the two
plt.figure(figsize=(15, 5))
plt.plot(ts[time_start:time_end], c='b', label='True time series')
plt.plot(y_pred_lag_1[time_start:time_end], c='r', ls=':', label='Lag 1 Forecast')
plt.xlabel('Time')
plt.ylabel('Electricity demand (kWh)')
plt.legend();
```

As was designed, the forecast is just one lag behind the true time series which provides good forecasts because of how small the time intervals are and as we observed before $X_t$ and $X_{t-1}$ are *highly* correlated. 

Despite how good this model does look like, it should be taken with a huge grain of salt for a few reasons:
- This model doesn't provide any sense of feature importance since it is non-parametric.
- More importantly, given the values $\{X_1, X_2, \dots, X_t\}$ it <b>only</b> allows us to predict the *next* value $X_{t+1}$. But to be generalizable and at all useful, we should be able to forecast well into the future (days, weeks or months)! 

In fact, let's say we wanted to predict hourly electricity demand for the next half a year. Then an analogous naive model to the $\hat{X}_{t+1} = X_t$ approach would be given $\{X_1, \dots, X_t\}$ predict $\hat{X}_{t+1}$ with $X_{t+1-k}$ where $k$ is the lag corresponding to one half a year ago.

### Exponential time smoothing

We start our first time series based forecasts with exponential smoothing [source here](https://otexts.com/fpp2/ses.html). We start with simple exponential time smoothing where we have that the prediction of $X_t$ is:<br>

$\hat{X}_t = \alpha X_{t-1} + (1-\alpha) \hat{X}_{t-1}$ <br>

This model is called exponential since we observe an exponential recursion with the $\alpha$ since <br>

$\hat{X}_{t-1} = \alpha X_{t-2} + (1-\alpha) \hat{X}_{t-2}$ <br>
$\hat{X}_{t-2} = \alpha X_{t-3} + (1-\alpha) \hat{X}_{t-3}$ <br>
$\dots$ <br>
$\hat{X}_2 = \alpha X_1 + (1-\alpha)\hat{X}_1$ <br>
$\hat{X}_1 = \alpha X_0 + (1-\alpha)\hat{X}_0$ <br>

where $\hat{X}_0$ and $\alpha$ are parameters that need to be learned from the model. This model assumes no trend or seasonality, it simply predicts the next observation by weighting the previous observation and forecast. For any following lag $\hat{X}_{t+h} = \hat{X}_t = \alpha X_{t-1} + (1-\alpha) \hat{X}_{t-1}$. So although we don't expect good results, we are interested in seeing if this can beat our baseline estimators (especially $\hat{X}_{t+h} = \hat{X}_{t+h-k}$ for some fixed $k$).<br>
We also give this model in component form since this will be used to represent more complex models:<br>
Forecast component $\hat{X}_{t+h} = l_t $ <br>
Smoothing component $l_t = \alpha X_{t-1} + (1 - \alpha) l_{t-1}$

```python
# Will need a wrapper class for the ts models to act like sklearn fit(X, y) and predict(X)
class TimeSeriesWrapper:
    """
    A class to wrap the statsmodel models so that we can call fit and predict.

    ...

    Attributes
    ----------
    model : 
      The statsmodel model to wrap.
    kwargs :
      Set of keyword arguments used for the statsmodel instance call.

    Methods
    ----------
    fit(X, y=None):
      Fits X to the model.
    predict(X):
      Predict the value for X.
    """

    def __init__(self, model, **kwargs):
      """
      Construct attributes for wrapper instance.

      Parameters
      ----------
      model : 
        Statsmodel model to wrap.
      kwargs:
        Key word arguments used for the model.
      """
      self.model = model
      self.kwargs = kwargs

    def fit(self, X, y=None):
      """
      Fit the model using the statsmodel syntax by passing X and kwargs and 
      then calling fit.

      Parameters
      ----------
      X : Series
        We expect this to be a series object
      y : 
        The y value is not used but necessary for the use of the cross-valid
        function.

      Returns
      -------
      None
      """
      # Suppress any convergence warnings
      with warnings.catch_warnings():
          warnings.filterwarnings("ignore")
          self.model_fit = self.model(X, **self.kwargs).fit()
        
    def predict(self, X):
      """
      Predict for values of X using the statsmodel pridct syntax.

      Parameters
      ----------
      X : Series
        We expect this to be a series object that we want to predict

      Returns
      -------
      y_pred : 
        The predicted values for X.
      """
      return self.model_fit.predict(X.index[0], X.index[-1])
```

```python
# CV error with simple exponential smoothing
n_splits = 10 # As stated before use 10 folds to keep the half-year consistency
model_simple_ets = TimeSeriesWrapper(ExponentialSmoothing,
                                     initialization_method="estimated")
mse_simple_ets, rmse_simple_ets = rolling_window_time_series(model_simple_ets, 
                                                            n_splits,
                                                            ts_trn,
                                                            ts_trn)
```

```python
print(f"The simple exponential smoothing validation mse is {round(mse_simple_ets, 2)} and "
      f"the validation rmse is {round(rmse_simple_ets, 2)}")
```

This is worse than any of our baselines! This might be because the constant estimate that the simple exponential smoothing gives biases the prediction more than the mean or median estimate. We will next try models that have a trend component: <br>

Forecast component $\hat{X}_{t+h} = l_t + h b_t $ <br>
Level component $l_t = \alpha X_{t-1} + (1 - \alpha) (l_{t-1} + b_{t-1})$ <br>
Trend compoenent $b_t = \beta (l_t - l_{t-1}) + (1 - \beta)b_{t-1} $ <br>

The parameters $\alpha, \beta$ are learned by the model through a non-linear optimization of the SSE. So we see now that forecasts are no longer the same for every $h$, but are a linear function of $h$. We can also add a damping effect which dampens the effect of the trend component to being constant for large $h$. This is more consistent with our data and real life data in general since it is rare to see a constant linear trend which does not change for values far away from the training set. The damped trend coefficient $\phi$ is also learned by the model. We try both for thoroughness. 

```python
# Trend not damped
model_trend = TimeSeriesWrapper(ExponentialSmoothing,
                                trend="add",
                                initialization_method="estimated")
mse_trend, rmse_trend = rolling_window_time_series(model_trend,
                                                  n_splits, 
                                                  ts_trn,
                                                  ts_trn)
```

```python
print(f"The simple exponential smoothing validation mse is {round(mse_trend, 2)} and "
      f"the validation rmse is {round(rmse_trend, 2)}")
```

```python
# Why is the trend model so bad?
# Retest this by training a sample and then predict for a validation set
trn_range = 1000
test = ExponentialSmoothing(ts_trn[:trn_range], trend="add").fit()
# The model predicts a negative trend...
ts_pred =  test.predict(trn_range, trn_range*2) 
print(ts_pred)
```

```python
# Trend damped
model_trend_damped = TimeSeriesWrapper(ExponentialSmoothing,
                                trend="add",
                                damped_trend=True,
                                initialization_method="estimated")
mse_damped_trend, rmse_damped_trend = rolling_window_time_series(model_trend_damped,
                                                                 n_splits, 
                                                                 ts_trn,
                                                                 ts_trn)
```

```python
print(f"The simple exponential smoothing validation mse is {round(mse_damped_trend, 2)} and "
      f"the validation rmse is {round(rmse_damped_trend, 2)}")
```

```python
# The damped trend is better than the non-damped, but it's still bad.
# Same experiment
trn_range = 1000 
test = ExponentialSmoothing(ts_trn[:trn_range], trend="add", damped_trend=True).fit()
# The model also predicts a negative trend! It is just less prnounced because of damping
test.predict(trn_range, trn_range*2) 
# This suggests that we should also account for seasonality, which we already knew.
```

We can also add a seasonal component to exponential smoothing with or without a trend component (with or without damping). 

```python
# Try different seasonalities
seasonal_dict = {'day': 24, 'week': 24*7, 'monthly': 30*24 }
seasonal_mse_dict, seasonal_rmse_dict = dict(), dict()
for k, v  in seasonal_dict.items():
    # Use 10-fold CV for our half-year objective
    n_splits = 10
    
    model_seasonal = TimeSeriesWrapper(ExponentialSmoothing,
                                  seasonal="add",
                                  seasonal_periods=v,
                                  initialization_method="estimated")
    seasonal_mse_dict[k], seasonal_rmse_dict[k] = rolling_window_time_series(model_seasonal,
                                                                            n_splits,
                                                                            ts_trn,
                                                                            ts_trn)
```

```python
# Report findings
for k, v in seasonal_dict.items():
    print(f"The exp. smoothing with period={k} has validation mse {round(seasonal_mse_dict[k], 2)} "
          f" and validation rmse is {round( seasonal_rmse_dict[k], 2)}")
```

We see that once we introduce seasonality, the exponential smoothing fairs much better. The weekly period does better than our baseline.

```python
# Now try seasonality + trend with and without damping
trend_damp = [True, False]
seastrend_mse_dict = dict()
seastrend_rmse_dict = dict()
for damp in trend_damp:
    for k, v  in seasonal_dict.items():
        # monthly 10-fold CV
        n_splits = len(ts_trn)//(24*30)
        if k == 'monthly':
            n_splits = 10
            
        model_seasonal = TimeSeriesWrapper(ExponentialSmoothing,
                                          trend="add",
                                          damped_trend=damp, 
                                          seasonal="add",
                                          seasonal_periods=v,
                                          initialization_method="estimated")
        key = f'period = {k}' + f' damped = {damp}'
        seastrend_mse_dict[key], seastrend_rmse_dict[key] = rolling_window_time_series(model_seasonal,
                                                                                     n_splits,
                                                                                     ts_trn,
                                                                                     ts_trn)
        print(f"Finished damp = {damp} and season = {k}")
```

```python
# Report findings
for k, v in seastrend_mse_dict.items():
    print(f"The exp. smoothing with config={k} has validation mse {round(seastrend_mse_dict[k], 2)} "
          f"and validation rmse is {round(seastrend_rmse_dict[k], 2)}")
```

### Arima and Sarima

We next turn to another time series based forecasting method: non-seasonal and seasonal ARIMA [source here](https://otexts.com/fpp2/arima.html). In exponential smoothing, forecasts were based on explicitly modeled the trend and seasonality of the data. In ARIMA models, models are based on the correlation between observations (this is where the auto-correlation plots and decomposition plots that we performed previously will be useful).

It is here that we introduce the notion of stationarity of a time series and the idea of using differencing to create a stationary time series. We say that a time series $\{X_1, X_2, \dots \}$ is strongly stationary if for every $n \in \mathcal{N}, h \in \mathcal{Z}$:<br>
$(X_1, X_2, \dots, X_n) \approx (X_{1+h}, X_{2+h}, \dots, X_{n+h})$ where $\approx$ means identically distributed.<br>
Clearly, this is unrealtistic to expect from most datasets. Especially electricity demand whose distribution may vary from one seasonal to another. So in general a (weakly) stationary time series refers to a time series $\{X_1, X_2, \dots \}$ such that: <br>
<ol>
<li>$E[X_t]$ is independent of time</li>
<li>$Cov(X_{t+h}, X_t)$ is independent of time for any lag $h$</li>
</ol>
What this is really saying is that there is no observable change in the behaviour of the time series from one point to another. Some time series may already be stationary. Others may not be. In this case, one technique to make the time series stationary is differencing. To talk about differecing we will define the following backshift operator $B$ as:<br>
$B^k X_t = X_{t-k}$<br>
What we do in differencing is take the original time series values $X_t$ and difference them with a particular lag. For instance 1-lag differencing is as follows:<br>
$(1 - B)X_t = X_t - X_{t-1}$<br>
The idea is it may be easier to model the difference of time series values rather than the time series itself.<br>
We can apply seasonal differencing where if we know there is a seasonal component that repeats at every lag $d$ then we can use the difference:<br>
$(1 - B^d)X_t = X_t - X_{t-d}$<br>
We apply some seasonal as well as time differences and plot the results. The ARIMA models <b>do this for us</b> in fact, but we try to get a sense of what will and won't work.

```python
# Use the shift operator to difference
ts_lag1 = ts_trn.shift(-1)
# X_t - X_{t-1}
plt.plot(ts - ts_lag1);
plt.title("Plot of $X_t - X_{t-1}$ versus time")
plt.ylabel("Electricity demand (kWh)")
plt.xlabel("Time")
```

```python
# Use the shift operator for seasonal differencing
# Apply them sequentially
# Day
ts_lag_day = ts_trn - ts_trn.shift(-24)
plt.plot(ts_lag_day)
plt.title("Plot of $X_t - X_{t-24}$ versus time")
plt.ylabel("Electricity demand (kWh)")
plt.xlabel("Time")
# Week
ts_lag_week = ts_lag_day - ts_lag_day.shift(-24*7) # 7*24 is a week's lag
plt.figure()
plt.plot(ts_lag_week)
plt.title("Plot of $X_t - X_{t- 7 * 24}$ versus time") 
plt.ylabel("Electricity demand (kWh)")
plt.xlabel("Time")
# Year
plt.figure()
ts_lag_year = ts_lag_week - ts_lag_week.shift(-24*365) # 365*24 is a year's lag
plt.plot(ts_lag_year)
plt.title("Plot of $X_t - X_{t- 365 * 24}$ versus time") 
plt.ylabel("Electricity demand (kWh)")
plt.xlabel("Time")
# Apply differencing
plt.figure()
ts_lagyear1 = ts_lag_year - ts_lag_year.shift(-1)
plt.plot(ts_lagyear1)
plt.title("Plot of $X_t - X_{t- 365 * 24}$ differenced again versus time") 
plt.ylabel("Electricity demand (kWh)")
plt.xlabel("Time");
```

We see that differencing even for a daily difference seems to remove some of the trend and cause the series to seem ''more stationary'' than before.

We use a the Dikey-Fuller test (unit root test) where the null hypothesis $H_0$ is that there is some underlying time component in the time series (i.e. non-stationary) and the alternative $H_a$ is that there is no underlying time component (i.e. stationary) to confirm our intuition [source here for unit root tests](https://otexts.com/fpp2/stationarity.html). The results are surprising:

```python
# Get the result of the original time series
result = adfuller(ts_trn)
print(f"p-value of test = {result[1]}") # reject 
```

```python
# Get the result for the time 1 difference
result = adfuller(ts_lag1.dropna())
print(f"p-value of test = {result[1]}") # reject
```

```python
# Get the result for seasonal differencing
result = adfuller(ts_lag_year.dropna())
print(f"p-value of test = {result[1]}") # reject
```

```python
# Get the result for seasonal + lag differencing
result = adfuller(ts_lagyear1.dropna())
print(f"p-value of test = {result[1]}") # reject
```

In all cases we would <b>*REJECT*</b> $H_0$ in favor of $H_a$ concluding erroneously that the data is stationary when there is a clear seasonal component. Part of the reasons why the statistical test fails can be found [here](https://stats.stackexchange.com/questions/225087/seasonal-data-deemed-stationary-by-adf-and-kpss-tests) if you are curious. A very rough explanation is that the test is more suited for detecting trend stationarity rather than seasonal stationarity. We know that we need to account for seasonality which is why we will try a Seasonal ARIMA with difference for day, week, month and year.

The reason why we need stationarity is because the ARIMA models which we will train next depend on the data being stationary. The Autoregressive Integrated Moving Average (ARIMA) model combines a linear combination of past time series values (AR), a differencing (I) and a combination of past forecasting errors [source here](https://otexts.com/fpp2/non-seasonal-arima.html). The non-seasonal ARIMA has three hyperaparameters:<br>
<ul>
    <li>$p$: the order of the AR part i.e. until which previous observation do we take a linear combination</li>
    <li>$d$: the degree of differencing</li>
    <li>$q$: the order of the MA part i.e. until which forecast error do we take</li>
</ul>
Together this gives the ARIMA$(p,d,q)$ model which has the following form:<br>
$\underbrace{(1-\phi_1 B-\phi_2 B^2 - \dots \phi_p B^p)}_{\text{AR(p)}} \underbrace{(1 - B)^d}_{\text{differencing $d$ times}} X_t = \underbrace{(1 + \theta_1 B + \theta_2 B^2 + \dots + \theta_q B^q)}_{\text{MA(q)}} Z_t$ <br>
Where $Z_t$ is white noise. This model works for non-seasonal time series. So we expect it to perform very poorly. Because of this, we will only consider a few hyperparameter settings $p=\{0,1,2\}, d = 1, q=\{0, 1, 2\}$ in the training.

```python
# Consider the set of possible models
p_max, d_max, q_max = 2, 1, 2
p_arr = np.array(range(p_max+1), dtype=np.int32)
d_arr = np.array(range(d_max+1), dtype=np.int32)
q_arr = np.array(range(q_max+1), dtype=np.int32)
p_grid, d_grid, q_grid = np.meshgrid(p_arr, d_arr, q_arr, indexing='ij')
hyperparam_arr = np.column_stack([p_grid.ravel(), d_grid.ravel(), q_grid.ravel()])
rmse_grid = np.empty_like(p_grid)
mse_grid = np.empty_like(p_grid)
```

```python
for p, d, q in hyperparam_arr:
    n_splits = 10 # ARIMAS are much more expensive to train that exp smoothin
    # reduce the number of folds
    model_arima = TimeSeriesWrapper(ARIMA, order=(p, d, q))
    # d is always 1 so d-1 is always 0
    mse_grid[p, d, q], rmse_grid[p, d, q] = rolling_window_time_series(model_arima,
                                                                          n_splits,
                                                                          ts_trn,
                                                                          ts_trn)
    print(f"Finished ARIMA({p}, {d}, {q})")
```

```python
for p, d, q in hyperparam_arr:
    print(f"ARIMA({p},{d},{q}) validation mse = {round(mse_grid[p, d, q], 2)} and "
          f"rmse = {round(rmse_grid[p, d, q], 2)}")
```

The non-seasonal ARIMA does not do better than our current optimal model because it doesn't consider seasonality. A seasonal ARIMA has the hyperparameters from the non-seasonal ARIMA $(p, d, q)$ as well as hyperparamters for its seasonal component $(P, D, Q)_m$ where $m$ is the number of observations in a period. If we let $\phi_p (z) = 1 - \sum_{i=1}^p \phi_i z^i$ and $\theta_q (z) = 1 + \sum_{i=1}^q \theta_i z^i$, then we can represent the seasonal ARIMA or SARIMA $(p, d, q)$ $(P, D, Q)_m$ as: <br>
$\phi_p(B) \phi_P(B^m) (1-B)^d (1-B)^D X_t = \theta_q(B) \theta_Q(B^m) Z_t$ <br>
For our purposes, we will not do a hyperparameter search over the non-seasonal hyperparamters again. We will we would like to use the optimal hyperparameters from before and then do a hyperparamter search for $P=\{0, 1, 2\}, D = \{0, 1\}, Q = \{0, 1, 2\}, m=\{24, 7\times24, 30\times 24\}$. However this will be computationally prohibitive because of how the SARIMA is computed. (We are not using this as an excuse to be lazy, we tried this on Google Colab with 25 GB of ram (!) and still got memory overflow). For the sake of completeness we provide the code that we would have used if we had a supercomputer.

```python
# Get the optimal hyp parameters from mse_grid
order_arima = np.unravel_index(np.argmin(mse_grid, axis=None), mse_grid.shape)
order_arima
```

```python
# Hyp parameter search
n_splits = 10
m_arr = np.array([4, 7*24, 30*24], dtype=np.int32) # m is the added seasonality component
p_grid, d_grid, q_grid, m_grid = np.meshgrid(p_arr, d_arr, q_arr, m_arr, indexing='ij')
hyperparam_arr = np.column_stack([p_grid.ravel(), d_grid.ravel(), q_grid.ravel(), m_grid.ravel()])
mse_grid_sarima, rmse_grid_sarima = np.empty_like(p_grid), np.empty_like(p_grid)   
hyperparam_arr
```
